# CNN


<br/><br/>

$$
\text{Input vector: } \\
\vec{x} = \begin{bmatrix}
           3 \\
           -2 \\
           \end{bmatrix} \\
$$

<br/><br/>

$$
\text{Neuron weights and bias :} \\
\vec{W} = \begin{bmatrix}
           \frac{1}{6} \\
           - \frac{3}{4} \\
           \end{bmatrix}
\quad \vec{b} = -\frac{3}{2}
$$

<br/><br/>

![](https://git.scc.kit.edu/vy9905/ml2images/-/raw/main/FF.drawio.svg)


<center> <h1>Feed Forward Step</h1> </center>


$$
\vec{z}\quad =\quad  (\vec{x}_1 * \vec{w}_1) + (\vec{x}_2 * \vec{w}_2) + (\vec{b})  \quad=\quad (3 * \frac{1}{6}) + (-2 * -\frac{3}{4}) + (-\frac{3}{2}) \quad = \quad \frac{1}{2} + \frac{3}{2} -\frac{3}{2} \quad = \quad \frac{1}{2}
$$

$$\vec{a} \quad = \quad Relu(\vec{z}) \quad = \quad max(0, \frac{1}{2}) \quad = \quad \frac{1}{2}$$


<center> <h1>Backpropagation</h1> </center>


$$
\text{We use a modified MSE loss function: } \\
J(\theta) = \frac{1}{2} * (\vec{y} - \vec{a})^2 \\
\text{and the target: } \vec{y} = \frac{3}{4} \\
$$

<br/><br/>

$$
\text{To calculate the loss, we insert the the calculated $\vec{a}$ and the target $\vec{y}$}$$


$$

J(\theta) = \frac{1}{2}_(\vec{y}-(ReLU((\vec{x_1}_\vec{w_1})+(\vec{x_2}\*\vec{w_2})+\vec{b}))^2$$

$$J(\theta) = \frac{1}{2}*(\frac{3}{4}-\frac{1}{2})^2 =\frac{1}{32}$$

$$
\text{Remember the chain rule: } \quad f(x) = g(h(x)) \quad \rightarrow \quad f'(x) = g'(h(x)) * h'(x)
$$

<br/><br/>

$$
\frac{∂J(\theta)}{∂\vec{w}_1} \quad = \quad \frac{∂J(\theta)}{∂\vec{a}} * \frac{∂\vec{a}}{∂\vec{z}} * \frac{∂\vec{z}}{∂\vec{w}_1}
$$


<center> <h3>Calculate gradient for each component:</h3> </center> 
<br/><br/>

<center> <h3>Partial derivative of loss $J(\theta)$ with regards to activation function $a$ :</h3> </center>

$$
J(\theta) = \frac{1}{2}(y-a)^2
$$

<center> 
$$
\frac{∂J(\theta)}{∂\vec{a}} =   \frac{1}{2} * 2 * (\vec{y} - \vec{a}) * (-1) = (\vec{a} - \vec{y})
$$
<br/><br/>
    
<center> <h3>Partial derivative of activation function $a$ with regards to neuron-output $z$ :</h3> </center>

<img src="https://git.scc.kit.edu/vy9905/ml2images/-/raw/main/ReLU.png" alt="relu" width="400"/>
    
<br/><br/>

<center> $a = ReLU(z)$</center>

   <center>$z=\frac{1}{2} > 0$ </center> 
$$
 \frac{∂\vec{a}}{∂\vec{z}} = 1 \quad \text{(ReLU's derivation is either 1 or 0)}
$$
<br/><br/>
<center> <h3>Partial derivative of neuron-output $z$ with regards to weight $w_1$ :</h3> </center>

$$
\vec{z} = \vec{w_1}*\vec{x_1}+\vec{w_2}*\vec{x_2}+\vec{b}
$$

$$
\frac{∂\vec{z}}{∂\vec{w}_1} = \vec{x}_1
$$

<br/><br/>

<center> <h3>Putting all together: Partial derivative of Loss $J(\theta)$ with regards to weight $w_1$</h3> </center>

$$
\frac{∂J(\theta)}{∂\vec{w}_1} \quad = \quad \frac{∂J(\theta)}{∂\vec{a}} * \frac{∂\vec{a}}{∂\vec{z}} * \frac{∂\vec{z}}{∂\vec{w}_1} = \quad (\vec{a} - \vec{y}) * 1 * \vec{x}_1 \quad = \quad (\frac{1}{2} - \frac{3}{4}) * 1 * 3 = - \frac{3}{4}
$$


<center> <h3>Weight Update:</h3> </center>

$$
\text{Update weights with learning rate } \eta \\
\Delta \vec{W^n} = - \eta * \frac{∂J(\theta)}{∂\vec{W^n}} \\
\vec{W^{n+1}} = \vec{W^{n}} + \Delta \vec{W^n}
$$


$$
\text{Reminder: }\vec{w_1^0}=\frac{1}{6} = \frac{10}{60}. \text{ Calculate weight update given the learning rate } \eta = \frac{1}{15} \\
\vec{w}_1^1 = \frac{1}{6} + \left(-\frac{1}{15}* -\frac{3}{4} \right) = \frac{13}{60}
$$


<center> <h3>Do the same for $w_2$:</h3> </center>

$$
\frac{∂J(\theta)}{∂\vec{w}_2} \quad = \quad \frac{∂J(\theta)}{∂\vec{a}} * \frac{∂\vec{a}}{∂\vec{z}} * \frac{∂\vec{z}}{∂\vec{
w}_2}
$$

<br/><br/>

$$
\frac{∂J(\theta)}{∂\vec{a}} \text{and} \frac{∂\vec{a}}{∂\vec{z}} \text{are identical to earlier}
$$

<br/><br/>

$$
\text{The only derivative that changes is } \frac{∂\vec{z}}{∂\vec{w}_2}
$$

<br/><br/>

$$
\vec{z} = \vec{w_1}*\vec{x_1}+\vec{w_2}*\vec{x_2}+\vec{b}
$$

$$
\frac{∂\vec{z}}{∂\vec{w}_2} = \vec{x}_2
$$

$$
\text{Reminder: }\vec{w_2} = -\frac{3}{4}=\frac{45}{60}, \quad \vec{x_2} = -2, \quad \vec{a}=\frac{1}{2}, \quad y=\frac{3}{4} \\
\frac{∂J(\theta)}{∂\vec{w}_2} \quad = \quad \frac{∂J(\theta)}{∂\vec{a}} * \frac{∂\vec{a}}{∂\vec{z}} * \frac{∂\vec{z}}{∂\vec{
w}_2} \quad = \quad (\vec{a} - \vec{y}) * 1 * \vec{x}_2 \quad = \quad (\frac{1}{2} - \frac{3}{4}) * 1 * - 2 = \frac{1}{2}
$$

<br/><br/>

$$
\vec{w}_2^{n+1} \quad = \quad \vec{w}_2^{n} + \left( - \eta * \frac{∂J(\theta)}{∂\vec{w}_2^n} \right) \quad = \quad -\frac{3}{4} + \left(-\frac{1}{15}* \frac{1}{2} \right) = - \frac{47}{60} \\
$$

<br/><br/>

$$
\text{Update } \vec{b} \\
\frac{∂J(\theta)}{∂\vec{b}} \quad = \quad \frac{∂J(\theta)}{∂\vec{a}} * \frac{∂\vec{a}}{∂\vec{z}} * \frac{∂\vec{z}}{∂\vec{b}} \quad = \quad (\vec{a} - \vec{y}) * 1 * 1 \quad = \quad (\frac{1}{2} - \frac{3}{4}) * 1 * 1 = -\frac{1}{4} \\
\vec{b}_{n+1} = \quad \vec{b}_{n} + \left( - \eta * \frac{∂E}{∂\vec{b}_n} \right) \quad = \quad -\frac{3}{2} + \left(-\frac{1}{15}* -\frac{1}{4} \right) = - \frac{89}{60}
$$


<center> <h3>New Feedforward with updated weights:</h3> </center>

$$
\vec{x} = \begin{bmatrix}
           3 \\
           -2 \\
           \end{bmatrix}
\quad
\vec{W} = \begin{bmatrix}
           \frac{13}{60} \\
           - \frac{47}{60} \\
           \end{bmatrix}
\quad \vec{b} = -\frac{89}{60}
$$

$$ \vec{a}\quad =\quad ReLU \left( (\vec{x}\_1 _ \vec{w}\_1) + (\vec{x}\_2 _ \vec{w}\_2) + (\vec{b}) \right) \quad=\quad Relu \left( (3 _ \frac{13}{60}) + (-2 _ -\frac{47}{60}) + (-\frac{89}{60}) \right) \quad = \frac{11}{15} = \quad 0.733$$

$$
\text{Recalculate the loss: Reminder before weight update: } a=\frac{1}{2} \quad J(\theta) = \frac{1}{32} \\
\text{Now: } J(\theta) = \frac{1}{2}*(\frac{3}{4} - \frac{11}{15})^2 = \frac{1}{7200}
$$

<center> <h3>Loss is now lower. The neural network's prediction is closer to the label $3/4$. Repeat until $a = y$</h3> </center>


<center> <h2>Dying ReLU</h2> </center>

$$
\text{Use original weights } \vec{W} = \begin{bmatrix}
           \frac{1}{6}\\
           -\frac{3}{4} \\
           \end{bmatrix} \quad \vec{b} = -\frac{3}{2} \text{ but new inputs: } \vec{x} = \begin{bmatrix}
           1\\
           1 \\
           \end{bmatrix}
$$

<br/><br/>

$$
\vec{z} = \frac{1}{6} * 1 -\frac{3}{4}*1 - \frac{3}{2} = -\frac{25}{12}\\
\vec{a} = ReLU(z) = ReLU(-\frac{25}{12}) = 0
$$

<br/><br/>

$$
\text{Calculate Gradients: } \\
\frac{∂J(\theta)}{∂\vec{w}_1} \quad = \quad \frac{∂J(\theta)}{∂\vec{a}} * \frac{∂\vec{a}}{∂\vec{z}} * \frac{∂\vec{z}}{∂\vec{
w}_1}
$$

$$
\frac{∂J(\theta)}{∂\vec{w}_2} \quad = \quad \frac{∂J(\theta)}{∂\vec{a}} * \frac{∂\vec{a}}{∂\vec{z}} * \frac{∂\vec{z}}{∂\vec{
w}_2}
$$

$$
\frac{∂J(\theta)}{∂\vec{b}} \quad = \quad \frac{∂J(\theta)}{∂\vec{a}} * \frac{∂\vec{a}}{∂\vec{z}} * \frac{∂\vec{z}}{∂\vec{
b}}
$$

$$ \vec{z} < 0 \rightarrow \frac{∂\vec{a}}{∂\vec{z}} = 0 \rightarrow \text{All Gradients are zero}$$

<center> <h3>Weight update with zero gradient:</h3> </center>

$$
\vec{w}_1^{n+1} \quad = \quad \vec{w}_1^{n} + \left( - \eta * \frac{∂J(\theta)}{∂\vec{w}_1^n} \right) \quad = \quad \frac{1}{6} + \left(-\frac{1}{15}*0 \right) = \frac{1}{6}
$$

$$
\vec{w}_2^{n+1} \quad = \quad \vec{w}_2^{n} + \left( - \eta * \frac{∂J(\theta)}{∂\vec{w}_2^n} \right) \quad = \quad -\frac{3}{4} + \left(-\frac{1}{15}*0 \right) = -\frac{3}{4}
$$

$$
\vec{b}^{n+1} \quad = \quad \vec{b}^{n} + \left( - \eta * \frac{∂J(\theta)}{∂\vec{b}^n} \right) \quad = \quad -\frac{3}{2} + \left(-\frac{1}{15}*0 \right) = -\frac{3}{2}
$$

<center> <h3>Weights stay the same, no learning is possible (for this input)</h3> </center>

# Loss Functions.

In Training we need Loss / Error Functions that define "how well our network is currently performing". We use backpropagation with gradient descent to minimize the loss and increase the performance of our network.

The most commonly used loss functions are Mean Squared Error and Crossentropy.

![](https://miro.medium.com/max/4800/1*Czxrg6IRDJQ3peBRls2dWA.webp)

- MSE is used for **regressive** tasks (e.g. how big is a bounding box / where is a bounding box located)
- Crossentropy is used for **classification** (e.g. which class is inside bounding box)


## MSE

$L_\text{MSE}=\frac{1}{n}\sum_{i=1}^n(\text{prediction}_i-\text{label}_i)^2$

Let's say we have Prediction

[2 3 4]

and label

[2 4 1]

The MSE is calculated as

$\frac{(2-2)^2 + (3-4)^2+(4-1)^2}{3}$
= $\frac{0^2 + (-1)^2+3^2}{3} = \frac{10}{3} $


```python
prediction = torch.tensor([2.0, 3.0, 4.0])
label = torch.tensor([2.0, 4.0, 1.0])
mse_criterion = nn.MSELoss()
mse_loss = mse_criterion(prediction, label)
print(mse_loss.item())
```

## Crossentropy Loss

$L_\text{CE}=-\sum_{i=1}^n\text{label}_i \cdot \text{log} (\text{prediction}_i)$

Softmax is always applied beforehand to prediction

$\text{softmax}(z_i) = \frac{e^{z_i}}{\sum_{k} e^{z_k}}$

E.g. we have $ \text{prediction} = [4 \quad 2 \quad 1]$ and $\text{label} = [1 \quad 0 \quad 0]$

Then Softmax calculates

$[ \frac{e^4}{e^4+e^2+e^1} \quad \frac{e^2}{e^4+e^2+e^1}  \quad \frac{e^1}{e^4+e^2+e^1}]$

$[0.844 \quad  0.114 \quad 0.042]$


```python
sm = nn.Softmax(dim=1)
prediction = torch.tensor([[4.0, 2.0, 1.0]])

softmaxed_pred = sm(prediction)
print(f"softmax of {prediction} is ", softmaxed_pred)
```

And the crossentropy loss is

$L_\text{CE}= -(1 \cdot \text{log}(0.844) + 0 \cdot \text{log}(0.114) + 0 \cdot \text{log}(0.042)) = -\text{log}(0.844) = 0.169$


```python
label = torch.tensor([[1.0, 0.0, 0.0]])
ce_criterion = nn.CrossEntropyLoss()  # This function automatically applies softmax
ce_loss = ce_criterion(prediction, label)  # use the original prediction [4, 2, 1]
print("Crossentropy loss is", ce_loss.item())
```

# **Parameters**


Let's say we want to classify images into 5 possible classes.
The images have size $32\times32$ and are in RGB.
For this classification we want to use 2 Convolutional Layer and one fully connected layer.

- The first convolutional layer has 16 filters, stride 2 and a 5$\times$5 kernel
- The second convolutional layer has 32 filters, stride 1, a 3$\times3$ kernel
  and padding to keep the same feature map resolution

Calculate the number of trainable parameters!


## Calculate parameters of convolutional Layers

Number of parameters per conv.layer can be calculated as follows:

$\text{number parameters} =\text{numberFilters} \times 
(\text{kernelWidth} \times \text{kernelHeight} \times \text{inputChannels} + 1)$

For the first convolutional layer we get:

$16 \cdot (5 \cdot 5 \cdot 3 + 1) = 1216 \text{ Parameters}$

For the second convolutional layer we get:

$32 \cdot  (3 \cdot 3 \cdot 16 + 1) = 4640 \text{ Parameters}$


## Calculate parameters of fully connected layer

The number of parameters in convolutional layers is independent of the input resolution. But for the fully connected layer we need to know the resolution of the incoming feature map.

We can calculate the resolution after a convolution with:

$\text{new resolution} = ⌊\frac{\text{prev feature map size} - \text{kernel size} }{stride} + 1⌋$

Dimension of first featuremap (original image):

$\quad 32\times32\times3$

Resolution after first convolutional layer:

$\quad ⌊\frac{32-5}{2}+1⌋ = ⌊14.5⌋ = 14$

Dimension of feature map after first convolutional layer:

$ \quad 14 \times 14 \times 16$

Second convolution uses padding to keep the resolution identical but increases number of filters. So the dimension after the second convolutional layer is:

$ \quad 14 \times 14 \times 32$

Now we can calculate the number of trainable parameters of the fully connected layer:

$\text{number parameters} = (\text{input features} + 1) \times \text{number neurons} $

which is:

$\quad (14 \cdot 14 \cdot 32 + 1) \cdot 5 = 31365$


So we have

1216 + 4640 + 31365 = 37221 parameter.


```python
model = nn.Sequential(
    nn.Conv2d(in_channels=3, out_channels=16, kernel_size=5, stride=2),
    nn.ReLU(),
    nn.Conv2d(
        in_channels=16,
        out_channels=32,
        kernel_size=3,
        stride=1,
        padding="same",
        padding_mode="zeros",
    ),
    nn.ReLU(),
    nn.Flatten(),
    nn.Linear(in_features=14 * 14 * 32, out_features=5),
    nn.Softmax(),
)


sum_of_trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print(f"this model has {sum_of_trainable_params} parameters")
```

Create Pytorch Dataset.
We use the normalization of ImageNet for our images and crop them to 224x224.
We use Batchsize 32.


```python
# Normal images have pixels in range [0,255]. It's oftentimes better to normalize them.
# There are different methods: Dividing by 255 creates a range of [0,1].
# Additionally you can calculat mean and std of dataset and then --> (image-mean)/std.
# Oftentime the means and std of ImageNet are used because we use ImageNet models

normalize = transforms.Normalize(
    mean=[0.485, 0.456, 0.406],  # mean of ImageNet
    std=[0.229, 0.224, 0.225],
)  # std of ImageNet

# specify transformations that should be applied to the image.
# (size-croppint, augmentation is done here)
data_transforms = {
    "train": transforms.Compose(
        [
            transforms.CenterCrop(224),  # crop every image to size 224x224
            transforms.ToTensor(),  # saves image as tensor (automatically divides by 255)
            normalize,
        ]
    ),  # (image-mean)/std.
    # sometimes we handle the training dataset differently than the validation/test dataset. E.g. if we used image augmentation
    # in training we don't want to use it in validation. Here it's the same
    "validation": transforms.Compose(
        [transforms.CenterCrop(224), transforms.ToTensor(), normalize]
    ),
}

# ImageFolder takes a file structure in which all images of class A are saved in folder A and all images of class B are saved in folder B
image_datasets = {  # create two datasets
    "train": datasets.ImageFolder(
        root="alienvspredator/train", transform=data_transforms["train"]
    ),
    "validation": datasets.ImageFolder(
        root="alienvspredator/validation", transform=data_transforms["validation"]
    ),
}

# Dataloaders are used to iterate over the dataset. They batch our data, shuffle our data, and use multiprocessing to speed up
# loading of data
dataloaders = {  # create two dataloaders.
    "train": torch.utils.data.DataLoader(
        image_datasets["train"], batch_size=32, shuffle=True, num_workers=0
    ),
    "validation": torch.utils.data.DataLoader(
        image_datasets["validation"], batch_size=32, shuffle=False, num_workers=0
    ),
}
```

Load Pretrained Model


```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
print("RUNNING ON ", device)  # should be cuda


def get_model_and_optimizer(pretrained=True):
    # load model architecture from pytorch and also return optimizer
    # if pretrained == true, then also load imagenet weights. otherwise use kaiming initialization

    if pretrained:
        # "DEFAULT" are weights trained on ImageNet
        model = models.efficientnet_v2_s(weights="DEFAULT").to(device)  # put it to GPU

        # We don't want to change the learned ImageNet weights. Disable gradient calculation for CNN
        for param in model.parameters():
            param.requires_grad = False
    else:
        # None --> Kaiming/He initialization
        model = models.efficientnet_v2_s(weights=None).to(device)

    # create a new classier
    # most models have a attribute classifier which symbolizes the last fully connected layer of a CNN
    # we overwrite the original ImageNet classifier
    model.classifier = nn.Sequential(
        nn.Linear(1280, 128),  # final feature map of CNN contains 1280 weights
        nn.ReLU(inplace=True),
        nn.Linear(128, 2),
    ).to(device)  # 2 classes

    # use AdamW and give it the trainable parameters
    if pretrained:
        ## only calculate gradients of the classifier
        optimizer = torch.optim.AdamW(
            model.classifier.parameters(), lr=5e-4, weight_decay=0.1
        )
    else:
        # calculate gradients for full network
        optimizer = torch.optim.AdamW(model.parameters(), lr=5e-4, weight_decay=0.1)

    return model.to(device), optimizer
```

```python
def validation(model):
    model.eval()
    with torch.no_grad():  # don't calculate gradients in validation
        running_corrects = 0  # used for accuracy calculation
        for inputs, labels in dataloaders["validation"]:
            inputs = inputs.to(device)
            labels = labels.to(device)
            outputs = model(inputs)
            _, predicted_class = torch.max(
                outputs, 1
            )  # _ = the highest prediction, --> predicted_class, which index hat the highest predicition
            running_corrects += torch.sum(predicted_class == labels)

        print(
            f"Validation accuracy: {running_corrects / len(image_datasets['validation']) * 100}%"
        )


def train_model(model, optimizer, num_epochs=5):
    start = time.time()
    criterion = nn.CrossEntropyLoss()

    for epoch in range(num_epochs):
        print(f"Epoch {epoch + 1}/{num_epochs}")
        print("-" * 10)

        running_loss = 0.0
        running_corrects = 0

        for inputs, labels in dataloaders["train"]:
            model.train()

            inputs = inputs.to(device)
            labels = labels.to(device)
            outputs = model(inputs)
            loss = criterion(outputs, labels)

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            running_loss += loss.item() * inputs.size(0)
            _, predicted_class = torch.max(outputs, 1)
            running_corrects += torch.sum(predicted_class == labels)

        epoch_loss = running_loss / len(image_datasets["train"])
        epoch_acc = running_corrects / len(image_datasets["train"])

        print(
            f"training epoch loss is = {epoch_loss}, Training accuracy is {epoch_acc}"
        )
        validation(model)

    print(f"This took {round(time.time() - start)} seconds")
    return model


pretr_model, optimizer = get_model_and_optimizer()
pretr_model = train_model(pretr_model, optimizer, num_epochs=5)
```

```python
def visualize_most_confident_images(model):
    a_valid = "alienvspredator/validation/alien/"
    p_valid = "alienvspredator/validation/predator/"
    validation_img_paths = [a_valid + x for x in sorted(os.listdir(a_valid))]
    validation_img_paths += [p_valid + x for x in sorted(os.listdir(p_valid))]

    img_list = []
    for img_path in validation_img_paths:
        i = Image.open(img_path)
        if i.mode == "RGB":
            img_list.append(i)
    with torch.no_grad():
        validation_batch = torch.stack(
            [data_transforms["validation"](img).to(device) for img in img_list]
        )
        prediction_logits_tensor = model(validation_batch)
        sm = nn.Softmax(dim=1)
        prediction_probs = sm(prediction_logits_tensor).cpu().data.numpy()

        alien_argmax = (-prediction_probs[:, 0]).argsort()[:5]
        pred_argmax = (-prediction_probs[:, 1]).argsort()[:5]

        uncertain = np.abs(prediction_probs[:, 0] - prediction_probs[:, 1])
        uncertain = uncertain.argsort()[:5]

    fig, axs = plt.subplots(3, 5, figsize=(20, 10))

    for idx, i in enumerate(alien_argmax):
        img = img_list[i]
        ax = axs[0, idx]
        ax.axis("off")
        ax.set_title(
            f"{100 * prediction_probs[i, 0]:.0f}% Alien, {100 * prediction_probs[i, 1]:.0f}% Predator"
        )
        ax.imshow(img)

    for idx, i in enumerate(pred_argmax):
        img = img_list[i]
        ax = axs[1, idx]
        ax.axis("off")
        ax.set_title(
            f"{100 * prediction_probs[i, 0]:.0f}% Alien, {100 * prediction_probs[i, 1]:.0f}% Predator"
        )
        ax.imshow(img)

    for idx, i in enumerate(uncertain):
        img = img_list[i]
        ax = axs[2, idx]
        ax.axis("off")
        ax.set_title(
            f"{100 * prediction_probs[i, 0]:.0f}% Alien, {100 * prediction_probs[i, 1]:.0f}% Predator"
        )
        ax.imshow(img)


visualize_most_confident_images(pretr_model)
```

```python
def rotate(model):
    img = "alienvspredator/validation/alien/0.jpg"
    img = Image.open(img)
    rotated_imgs = []
    for i in range(8):
        rotated_imgs.append(img.rotate(360 / 8 * i))
    fig, axs = plt.subplots(1, 8, figsize=(20, 10))

    for idx, img in enumerate(rotated_imgs):
        ax = axs[idx]
        ax.axis("off")
        ax.imshow(img)

    with torch.no_grad():
        rotated_batch = torch.stack(
            [data_transforms["validation"](i).to(device) for i in rotated_imgs]
        )

        prediction_logits_tensor = model(rotated_batch)
        sm = nn.Softmax(dim=1)
        prediction_probs = sm(prediction_logits_tensor).cpu().numpy()
        for i in range(len(prediction_probs)):
            print(
                f"Predicting: {prediction_probs[i, 0] * 100} % for Alien and {prediction_probs[i, 1] * 100} % for Predator"
            )


rotate(pretr_model)
```
